```python
def main():

        
    # define functions
    data=exploredata()
    computeRegression(data)
    
    

# The first function explores the data
def exploredata():
    import pandas as pd
    import matplotlib.pyplot as plt 

    # Read data set and load into a pandas data frame. Prompt user if file is not in the right location
    
    try:
        data= pd.read_csv('MedicalAppointment.csv')
    
        # check the first few lines of the data

        print(data.head(20))

        # check the last few rows of data

        data.tail()

        # Review data frame info
        data.info()

        # descriptions and summary statistics 
     
        print(data.describe())

        # Print the names of the columns in games.
        print(data.columns)

        # dimesnion of the data

        print(data.shape)

        # datatypes

        print(data.dtypes)

        # class distribution 
        print(data.groupby('Status').size())
    
        # Plot histogram of data
    
        # histograms 
        data.hist(sharex=False, sharey=False, xlabelsize=1, ylabelsize=1)
        plt.show()
    
        # Histogram of Age Variable-Continous variable
    
        data['Age'].hist(bins=10)
        plt.show()
    
        data_Status = pd.get_dummies(data['Status'])

        data_new = pd.concat([data, data_Status], axis=1)

        data_new.head()

        temp1=data['Gender'].value_counts(ascending=True)

        temp2=data['Diabetes'].value_counts(ascending=True)

    # Probability of showing-up for each Gender Type
        temp3=data.pivot_table(values='Status', index=['Gender'],aggfunc=lambda x: x.map({'Show-Up':1,'No-Show':0}).mean())

        print(temp1)
        print(temp2)
        temp1.plot(kind='bar')   
        plt.show()
    
        temp3.plot(kind='bar')
        plt.show()

        data['Gender'].value_counts()

        data['Diabetes'].value_counts()



        # Draw box-polts of continous variables

        #data.boxplot(column='Age')
        data.boxplot(column='Age', by='Status', ax=None, return_type='axes')

        plt.show()


    #data.boxplot(column='Age')
        data.boxplot(column='Age',  ax=None, return_type='axes')
        plt.show()
        
        # correlation matrix 
        fig = plt.figure() 
        ax = fig.add_subplot(111) 
        cax = ax.matshow(data.corr(), vmin=-1, vmax=1, interpolation='none') 
        fig.colorbar(cax) 
        plt.show()



    # Check for missing values in data set

        data.apply(lambda x: sum(x.isnull()),axis=0) 
    
        return (data)

    except FileNotFoundError:
        print("File not found. Please make sure file is in the right location")



    

    # The second function computes regression on the data. 

def computeRegression(data):
    import pandas as pd
    
    data = data[(data.Age >= 2) & (data.Age <= 100)]
# create dummy variables for status and Gender and create new columns for the dummy variables


    data_gender= pd.get_dummies(data['Gender'])

    data = pd.concat([data, data_gender], axis=1)

    data_Status = pd.get_dummies(data['Status'])

    data = pd.concat([data, data_Status], axis=1)

    

    # drop the columns that are duplicates or not needed

    data.drop(data.columns[[1, 5, 15, 17]],inplace=True, axis=1) 

    # Rename columns

    data.rename(columns = {'M':'Gender'}, inplace=True)
    data.rename(columns = {'ApointmentData':'AppointmentData',
                         'Alcoolism': 'Alchoholism',
                         'HiperTension': 'Hypertension',
                         'Handcap': 'Handicap'}, inplace = True)


    # Change all categorical variables to type category

    
    for col in ['DayOfTheWeek', 'Diabetes', 'Alchoholism', 'Hypertension', 'Handicap', 'Smokes', 'Scholarship','Tuberculosis','Sms_Reminder','Gender', 'Show-Up']:
        data[col] = data[col].astype('category')


    data.info()
    
    
    
    
    # Get all the columns from the dataframe.
    columns= data.columns.tolist()
    columns=columns[0:14]
    
    # Encode category variable for scikitlearn to process

    from sklearn.preprocessing import LabelEncoder
    var_cat = ['AppointmentRegistration','AppointmentData','DayOfTheWeek','Diabetes','Alchoholism','Hypertension','Handicap','Smokes','Scholarship','Sms_Reminder','Gender', 'Show-Up']
    le = LabelEncoder()
    for i in var_cat:
        data[i] = le.fit_transform(data[i])
    data.dtypes 
    # Store the variable we'll be predicting on.

    target= "Show-Up"

    # Import a convenience function to split the sets.

    from sklearn.cross_validation import train_test_split

    # Generate the training set.  Set random_state to be able to replicate results.

    train = data.sample(frac=0.7, random_state=1)

    # Select anything not in the training set and put it in the testing/validation set.

    test = data.loc[~data.index.isin(train.index)]

    # Print the shapes of both sets.
    print(train.shape)
    print(test.shape)
    
    # import libraries for models

    from sklearn.ensemble import RandomForestClassifier
    from sklearn.tree import DecisionTreeClassifier, export_graphviz
    from sklearn import metrics
    from sklearn.lda import LDA
    from sklearn import svm
    



    # Fit Logistic Regression

    from sklearn.linear_model import LogisticRegression 

    model = LogisticRegression()
    # Fit the model to the training data.
    model.fit(train[columns], train[target])

    print(model.coef_)

    #print(model.feature_importances_)


    #Make predictions on training set:
    predictions = model.predict(train[columns])
  
    #Print accuracy
    accuracy = metrics.accuracy_score(predictions,train[target])
    
    #Make predictions on Validation set:
    predictions_test = model.predict(test[columns])
  
    #Print accuracy
    accuracy_test = metrics.accuracy_score(predictions_test,test[target])

    # Print confusion matrix

    from sklearn.metrics import confusion_matrix

    matrix=confusion_matrix(predictions,train[target])

    print("Logistic Regression_Training : ",accuracy*100,"%")
    print("Logistic Regression_Test : ",accuracy_test*100,"%")

    print(matrix)
    
    # Write the output of the logistic regression to a text file
    
    outfile= open("logistic.txt", 'w')
    outfile.write(str(accuracy*100)+"\n")
    outfile.write(str(accuracy_test*100))
    outfile.close()

    
    # Print Classification Report


    from sklearn.metrics import classification_report 
    print (classification_report(train[target], predictions))


    # Fit Random Forest Classifier
    model = RandomForestClassifier(n_estimators=25, min_samples_split=25, max_depth=7, max_features=1)
    # Fit the model to the training data.
    model.fit(train[columns], train[target])
    
    #print("Maximum Likelihood Estimates of coefficients-Random Forest")
    
    #print(model.coef_)


    #Make predictions on training set:
    predictions = model.predict(train[columns])
  
    #Print accuracy
    accuracy = metrics.accuracy_score(predictions,train[target])
    
    #Make predictions on test set:
    predictions_test = model.predict(test[columns])
  
    #Print accuracy
    accuracy_test = metrics.accuracy_score(predictions_test,test[target])
    
    # display the relative importance of each attribute. This can be used for Feature Engineering
    print(model.feature_importances_)



    print("Random Forest_Training :",accuracy*100,"%")
    print("Random Forest_Test :",accuracy_test*100,"%")
    
    # write output to a text file
    
    outfile= open("Random.txt", 'w')
    outfile.write(str(accuracy*100)+"\n")
    outfile.write(str(accuracy_test*100))
    outfile.close()
    
    from sklearn.metrics import confusion_matrix

    matrix_RF=confusion_matrix(predictions,train[target])
    
    print('Random Forest Confusion Matrix')
    print(matrix_RF)


    # Fit Decision Tree Model
    
    model= DecisionTreeClassifier()


    # Fit the model to the training data.
    model.fit(train[columns], train[target])


    #Make predictions on training set:
    predictions = model.predict(train[columns])
  
    #Compute accuracy on training data
    accuracy = metrics.accuracy_score(predictions,train[target])
    
    #Make predictions on test/validation set:
    predictions_test = model.predict(test[columns])
  
    #Compute accuracy on validation data
    accuracy_test = metrics.accuracy_score(predictions_test,test[target])



    #Print Output:

    print("Decison Tree_Training :",accuracy*100, "%")
    print("Decison Tree_Test :",accuracy_test*100, "%")
    
    # Write the output of Decison Tree to text file
    
    outfile= open("Decision.txt", 'w')
    outfile.write(str(accuracy*100)+"\n")
    outfile.write(str(accuracy_test*100))
    outfile.close()
    
    # Fit Linear Discriminant Analysis
    model= LDA()


    # Fit the model to the training data.
    model.fit(train[columns], train[target])


    #Make predictions on training set:
    predictions = model.predict(train[columns])
  
    #Print accuracy
    accuracy = metrics.accuracy_score(predictions,train[target])
    
    #Make predictions on test set:
    predictions_test = model.predict(test[columns])
  
    #Print accuracy
    accuracy_test = metrics.accuracy_score(predictions_test,test[target])


    #Print Output:

    print("LDA_Training :",accuracy*100, "%")
    print("LDA_Test :",accuracy_test*100, "%")  
    
    outfile= open("lda.txt", 'w')
    outfile.write(str(accuracy*100)+ "\n")
    outfile.write(str(accuracy_test*100))
    outfile.close()
    
        
    

main()
```
